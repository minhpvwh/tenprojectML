```python
from shutil import copyfileobj
from six.moves import urllib
from sklearn.datasets.base import get_data_home
import os

def fetch_mnist(data_home=None):
    mnist_alternative_url = "https://github.com/amplab/datascience-sp14/raw/master/lab7/mldata/mnist-original.mat"
    data_home = get_data_home(data_home=data_home)
    data_home = os.path.join(data_home, 'mldata')
    if not os.path.exists(data_home):
        os.makedirs(data_home)
    mnist_save_path = os.path.join(data_home, "mnist-original.mat")
    if not os.path.exists(mnist_save_path):
        mnist_url = urllib.request.urlopen(mnist_alternative_url)
        with open(mnist_save_path, "wb") as matlab_file:
            copyfileobj(mnist_url, matlab_file)
```

```python
fetch_mnist()
```

```python
from sklearn.datasets import fetch_mldata
```

```python
mnist = fetch_mldata("MNIST original")
```

```python
mnist.data.shape
```

```python
# nó sẽ có 70000 bức ảnh
# mỗi dòng của nó sẽ là ảnh 28*28
```

```python
X_data = mnist.data
y_data = mnist.target
```

```python
y_data.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# draw one image
plt.imshow(X_data[10000].reshape(28,28))
```

```python
# muốn hiển thị 10 ảnh
nb_imgs = 10
# dùng subplots truyền vào số hàng , số cột và kích thước
fig, axes = plt.subplots(nb_imgs, nb_imgs, figsize=(15, 15))
# chúng ta có thể thấy hiện tại fig của chúng ta k có gì cả phải sử dụng một vòng for để đổ dữ liệu ảnh vào
for i in range(nb_imgs):
    for j in range(nb_imgs):
        axes[i][j].imshow(X_data[i*j].reshape(28, 28))
```

```python
# ta thấy dữ liệu đang chưa được suffle nên ta tiến hành trộn chúng
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
```

```python
# muốn hiển thị 10 ảnh
nb_imgs = 10
# dùng subplots truyền vào số hàng , số cột và kích thước
fig, axes = plt.subplots(nb_imgs, nb_imgs, figsize=(15, 15))
# chúng ta có thể thấy hiện tại fig của chúng ta k có gì cả phải sử dụng một vòng for để đổ dữ liệu ảnh vào
for i in range(nb_imgs):
    for j in range(nb_imgs):
        axes[i][j].imshow(X_train[i*j].reshape(28, 28))
```

```python
y_train_5 = (y_train == 5)
```

```python
y_train_5
```

```python
# binary classification
from sklearn.linear_model import SGDClassifier
```

```python
sgd = SGDClassifier(random_state=42)
```

```python
sgd.fit(X_train,y_train_5)
```

```python
y_pred = sgd.predict(X_test)
```

```python
#  nếu chúng ta muốn dánh giá được xem, mà k cần dùng đến tập test này thì chúng ta có thể dùng cross validation - chúng ta chia tập dữ liệu của chúng ta thành vài phần khác nhau , chúng ta giữ lại một phần trong số đó, ví dụ có 5 phần chúng ta giữ lại 1 phần trong số đó lấy 4 phần còn lại để chúng ta training, lặp đi lặp lại như thế
from sklearn.model_selection import cross_val_predict
# cv là có bn cross validation
y_train_pred =  cross_val_predict(sgd, X_train, y_train_5, cv=5)
```

```python
y_train_pred.shape
```

```python
# cross score
from sklearn.model_selection import cross_val_score
cross_val_score(sgd, X_train, y_train_5, cv=5, scoring='accuracy')
```

```python
#performance measure with confution matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_train_5, y_train_pred)
```

```python
# 49961 - True - negative - nếu nó sai nhưng đoán đúng nó là sai
# 1409 - False - positive - nếu nó đúng nhưng đoán sai nó là sai
# 1012 - False - nagetive - nếu nó sai nhưng đoán sai nó là đúng
# 3618 - True - positive - nếu nó đúng nhưng đoán đúng nó là đúng

# Precition not is number 5

precition_not_5 = 49961 / (49961 + 1409)
precition_not_5
```

```python
# Precition is number 5

precition_5 = 3618 / (3618 + 1012)
precition_5
```

```python
# recall not is number 5 
recall_not_5 = 49961 / (49961 + 1012)
recall_not_5
```

```python
# recall is number 5 
recall_5 = 3618 / (3618 + 1409)
recall_5
```

```python
# F1-score
def f1_score(precision, recall):
    return 2*precision*recall / (precision + recall)
f1_score(precition_not_5, recall_not_5)
```

```python
#print  classification report
from sklearn.metrics import classification_report
target_names= ['Not 5', 'Is 5']
print(classification_report(y_train_5, y_train_pred, target_names=target_names))
```
